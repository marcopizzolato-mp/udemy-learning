```python
# Libraries
import random
from blackjack_art import logo

# Functions
def get_cards():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    card = random.choice(cards)
    return card

def add_card(player_list):
    card = get_cards()
    if card == 11 and (sum(player_list) + card) > 21:
        card = 1
    player_list.append(card)

def print_int(list_user, list_dealer):
    print(f'Your cards: {list_user}, current score: {sum(list_user)}')
    print(f"Computer's first card: {list_dealer[0]}")
def print_fin(list_user, list_dealer):
    print(f'Your final hand: {list_user}, final score: {sum(list_user)}')
    print(f"Computer's final hand: {list_dealer}, final score: {sum(list_dealer)}")

def blackjack():
    print("Beginning")
    list_user = [20,1]
    list_dealer = []
    counter = 0
    another_play = 'y'
    while another_play == 'y': 

        # Set up the first round
        if counter == 0:
            for ii in range(0,2,1):
                # add_card(list_user)
                add_card(list_dealer)
        else:
            add_card(list_user)

        counter +=1
        # Print results
        print_int(list_user, list_dealer)
        print(list_user)
        print(list_dealer)
        if sum(list_user) > 21 or sum(list_user) == 21 or sum(list_dealer) > 21 or sum(list_dealer) == 21:
            another_play = 'n'
        else:
            # Ask user to play another round
            another_play = input("Do you want another card? Type 'y' or 'n': ")   

    
    
    if sum(list_dealer) < 17:
        add_card(list_dealer)
    
    if sum(list_dealer) == 21 and len(list_dealer) == 2:
        print("Computer wins with BlackJack!! ")
    elif sum(list_user) == 21 and len(list_user) == 2:
        print("You win with BlackJack!! ")
    elif sum(list_user) > 21: 
        print('You are above 21, you lose ')
    elif sum(list_dealer) > 21:
        print('Computer above 21, you win ')
    elif sum(list_user) == sum(list_dealer):
        print('This is a draw! No one loses or wins ')
    elif sum(list_user) > sum(list_dealer):
        print('You win ')
    elif sum(list_dealer) > sum(list_user):
        print('Computer wins.. sorry buddy ')
    else:
        print("BHO")
        
    print_fin(list_user, list_dealer)    
    wanna_play_blackjack()
        
def wanna_play_blackjack():
    wanna_play = input("Do you want to play a game of Blackjack? Type 'y' or 'n': ")
    if wanna_play == 'y':
        blackjack()
    else:
        print("You chose not to play.... BYE!")
```
