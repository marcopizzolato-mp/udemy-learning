```python
# Get random item from list of choices
def get_random(list_choices):
    random_choice = random.choices(list_choices)
    return random_choice[0]

# Assign the content of the dictionary to the string and print
def print_a(choice_dict):
    print(f"Compare A: {choice_dict['name']}, a {choice_dict['description']}, from {choice_dict['country']}, followers {choice_dict['follower_count']}.") 
    
# Assign the content of the dictionary to the string and print    
def print_b(choice_dict):
    print(f"Compare B: {choice_dict['name']}, a {choice_dict['description']}, from {choice_dict['country']}, followers {choice_dict['follower_count']}.") 
    
def compare_choices(user_guess, choice_a, choice_b):
    score_a = choice_a['follower_count']
    score_b = choice_b['follower_count']
    if score_a >= score_b:
        # The user selected 'a' and equal to a 
        return user_guess == 'a' 
    else:
        return user_guess == 'b' 

           
    # if (score_a > score_b and user_guess == "a") or (score_a < score_b and user_guess == "b"):
    #     print(f"You got it right! score_a {score_a} score_b {score_b}")
    #     increase_point()        
    #     go_ahead = True
    # elif (score_a < score_b and user_guess == "a") or (score_a > score_b and user_guess == "b"):
    #     print(f"You got it wrong! score_a {score_a} score_b {score_b}")
    #     go_ahead = False        
    # else:
    #     print("Uhhh... that's a draw! You go ahead.. but no points! ")
    #     go_ahead = True
    # return go_ahead

def continue_play():
    wp = input("Do you want to do another game? y/n: ").lower()
    if wp == "y":
        wanna_play = True
    elif wp == "n":
        wanna_play = False
    else:
        print("Hmm.. your choince is neither 'y' or 'n', type again.. ")
        continue_play()
    return wanna_play

def get_input():
    # Get input from user
    user_guess = input("Who has more followers? Type 'A' or 'B': ").lower()
    return user_guess
    
def increase_point():
    global points
    points = points + 1
    return points
    
```

```python
# Initiate point counter
points = 0
first_round = True
# Main game structure
wanna_play = True
while wanna_play:
    
    if first_round == True:
        choice_a = get_random(init_data)
        choice_b = get_random(init_data)
        first_round = False    
    # Logo
    # print(lowerhigher_art.logo)
    # Print choice
    print_a(choice_a)
    
    # Logo
    # print(lowerhigher_art.vs)
    # Print choice
    print_b(choice_b)

    # Loop till the choice is correct 
    user_guess = ''
    while user_guess != "a" and user_guess != "b":
        user_guess = get_input()
    result = compare_choices(user_guess, choice_a, choice_b)
    if result == False:
        print("--______________________________________________________________________--")
        wanna_play = continue_play()
    else: 
        print(points)
        print("Hurraayy !!")
        # Swap the items
        if user_guess == 'a':
            choice_b = get_random(init_data)
        else:
            choice_a = choice_b.copy()
            choice_b = get_random(init_data)


        
clear_output()    
```
