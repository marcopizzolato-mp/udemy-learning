```python
import random

def guess_number():
    # Generate the number beteen 1 and 100
    num = round(random.random(),2)
    num = int(num*100)
    num = 10
    # Select the level
    level = input("At which level do you want to play?? hard/easy ")
    level_dict = {'hard':5,'easy':10}
    level_rounds = level_dict[level]

    terminate_loop = False
    while level_rounds >= 1 and terminate_loop == False:
        print(f"You have {level_rounds} attempts left")
        level_rounds -= 1

        # Get the input 
        number = int(input("Make a guess: "))

        if number == num:
            print("Hurray, you win! ")
            terminate_loop = True
        elif number < num:
            print("Number is too low, shoot up! ")
        elif number > num:
            print("Number is too high, shoot down! ")
    
    if level_rounds == 0:
        print("Sorry, you are out of chances! ")

    wanna_play = input("Do you want to play again? y/n ")
    if wanna_play == "y":
        guess_number()
    else:
        print("You chose to go away... BYE! ")

guess_number()
```

```python
#Function to check user's guess against actual answer.
def check_answer(guess, answer, turns):
    if guess > answer:
        print("Too high.")
        return turns - 1
    elif guess < answer:
        print("Too low.")
        return turns - 1
    else:
        print(f"You got it! The answer was {answer}.")
```
